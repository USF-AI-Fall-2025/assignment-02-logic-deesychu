```python
# Author: Adrienne Dominique Sy Chu

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from DataInvestigator import DataInvestigator

# Load the gallstone dataset
df = pd.read_csv("gallstone.csv")

# Create DataInvestigator instance
di = DataInvestigator(df)
```

```python
# --- Q1: Gender Encoding ---

# Gender count plot
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=df, hue="Gender", palette="pastel", legend=False)
plt.title("Gender Encoding Distribution")
plt.xlabel("Gender Code")
plt.ylabel("Count")
plt.show()

# Age distribution by Gender
plt.figure(figsize=(7,5))
sns.boxplot(x="Gender", y="Age", data=df, hue="Gender", palette="pastel", legend=False)
plt.title("Age Distribution by Gender Code")
plt.xlabel("Gender Code")
plt.ylabel("Age")
plt.show()

# Check baseline gender using your DataInvestigator class
print("Baseline gender (most frequent):", df['Gender'].mode()[0])
print(df['Gender'].value_counts())
```

```python
# Q2: Glucose distribution

# Histogram with KDE to show distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Glucose'], kde=True, bins=30, color="skyblue")
plt.title("Glucose Distribution")
plt.xlabel("Glucose")
plt.ylabel("Count")
plt.show()

# Boxplot to identify outliers
plt.figure(figsize=(7,4))
sns.boxplot(x=df['Glucose'], color="lightgreen")
plt.title("Glucose Outliers")
plt.xlabel("Glucose")
plt.show()

# Violin plot to visualize distribution shape and spread
plt.figure(figsize=(7,5))
sns.violinplot(x=df['Glucose'], color="lightcoral")
plt.title("Glucose Distribution (Violin Plot)")
plt.xlabel("Glucose")
plt.show()

# Summary statistics
print("Glucose Summary:")
print(df['Glucose'].describe())
```
