```python
# Author: Adrienne Dominique Sy Chu

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from DataInvestigator import DataInvestigator

# Load the gallstone dataset
df = pd.read_csv("gallstone.csv")

# Create DataInvestigator instance
di = DataInvestigator(df)
```

```python
# --- PART 2 QUESTIONS ---
```

```python
# Q1: Gender encoding (0 or 1?)
# Shows two bars â€” one for 0, one for 1.

plt.figure(figsize=(6,4))
sns.countplot(x=df['Gender'])
plt.title("Gender Encoding Distribution")
plt.show()

print("Baseline gender (most frequent):", di.baseline(df.columns.get_loc('Gender')))
```

```python
# Q2: Glucose distribution
# Shows a histogram with a smooth curve on top

plt.figure(figsize=(8,5))
sns.histplot(df['Glucose'], kde=True, bins=30)
plt.title("Glucose Distribution")
plt.xlabel("Glucose")
plt.ylabel("Count")
plt.show()


# Boxplot to check for outliers 
# Shows the middle range of glucose values.
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Glucose'])
plt.title("Glucose Outliers")
plt.show()
```
